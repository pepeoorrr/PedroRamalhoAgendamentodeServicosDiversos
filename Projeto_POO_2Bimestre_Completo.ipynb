## PARTE 1 ‚Äì Classe Principal e Objetos

**Tarefas:**
- Definir e criar a classe principal com pelo menos 3 atributos.
- Implementar o m√©todo construtor (`__init__`).
- Criar pelo menos 2 objetos da classe principal.
- Implementar um m√©todo de exibi√ß√£o de dados.

**C√≥digo:**

```python
class AgendamentoServico:
    # Definindo o construtor com 3 atributos principais
    def __init__(self, nome_cliente, tipo_servico, data_agendamento):
        self.nome_cliente = nome_cliente      # Nome do cliente
        self.tipo_servico = tipo_servico      # Tipo de servi√ßo agendado
        self.data_agendamento = data_agendamento  # Data do agendamento

    # M√©todo para exibir os dados do agendamento
    def exibir_dados(self):
        print(f"Cliente: {self.nome_cliente}")
        print(f"Servi√ßo Agendado: {self.tipo_servico}")
        print(f"Data do Agendamento: {self.data_agendamento}")
        print("-" * 30)

# Criando objetos (inst√¢ncias) da classe AgendamentoServico
agendamento1 = AgendamentoServico("Maria Silva", "Corte de Cabelo", "2025-05-10")
agendamento2 = AgendamentoServico("Jo√£o Oliveira", "Limpeza de Pele", "2025-05-12")

# Exibindo os dados dos agendamentos
agendamento1.exibir_dados()
agendamento2.exibir_dados()
```

## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
class AgendamentoServico:
    def __init__(self, nome_cliente, tipo_servico, data_agendamento):
        self.nome_cliente = nome_cliente
        self.tipo_servico = tipo_servico
        self.data_agendamento = data_agendamento
        self.status = "Agendado"

    def exibir_dados(self):
        print(f"Cliente: {self.nome_cliente}")
        print(f"Servi√ßo Agendado: {self.tipo_servico}")
        print(f"Data do Agendamento: {self.data_agendamento}")
        print(f"Status: {self.status}")
        print("-" * 30)

    # M√©todo de a√ß√£o 1
    def remarcar(self, nova_data):
        self.data_agendamento = nova_data
        print(f"Agendamento de {self.nome_cliente} foi remarcado para {self.data_agendamento}.")

    # M√©todo de a√ß√£o 2
    def cancelar(self):
        self.status = "Cancelado"
        print(f"Agendamento de {self.nome_cliente} foi cancelado.")

# Subclasse com heran√ßa
class AgendamentoServicoResidencial(AgendamentoServico):
    def __init__(self, nome_cliente, tipo_servico, data_agendamento, endereco):
        super().__init__(nome_cliente, tipo_servico, data_agendamento)
        self.endereco = endereco  # Novo atributo

    # Novo m√©todo exclusivo da subclasse
    def confirmar_endereco(self):
        print(f"Servi√ßo residencial ser√° realizado no endere√ßo: {self.endereco}")

# Testes
agendamento1 = AgendamentoServico("Carlos Souza", "Manuten√ß√£o de Ar", "2025-06-01")
agendamento1.exibir_dados()
agendamento1.remarcar("2025-06-10")

agendamento2 = AgendamentoServicoResidencial("Luciana Alves", "Dedetiza√ß√£o", "2025-06-03", "Av. Central, 456")
agendamento2.exibir_dados()
agendamento2.confirmar_endereco()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
class AgendamentoServico:
    def __init__(self, nome_cliente, tipo_servico, data_agendamento):
        self.__nome_cliente = nome_cliente  # Atributo privado
        self.tipo_servico = tipo_servico
        self.data_agendamento = data_agendamento
        self.status = "Agendado"

    # Getter
    def get_nome_cliente(self):
        return self.__nome_cliente

    # Setter
    def set_nome_cliente(self, novo_nome):
        if novo_nome.strip() != "":
            self.__nome_cliente = novo_nome
        else:
            print("Nome inv√°lido!")

    def exibir_dados(self):
        print(f"Cliente: {self.__nome_cliente}")
        print(f"Servi√ßo Agendado: {self.tipo_servico}")
        print(f"Data do Agendamento: {self.data_agendamento}")
        print(f"Status: {self.status}")
        print("-" * 30)

    def remarcar(self, nova_data):
        self.data_agendamento = nova_data
        print(f"Agendamento de {self.__nome_cliente} foi remarcado para {self.data_agendamento}.")

    def cancelar(self):
        self.status = "Cancelado"
        print(f"Agendamento de {self.__nome_cliente} foi cancelado.")

    # M√©todo com l√≥gica interna (abstra√ß√£o)
    def verificar_agendamento(self):
        if self.status == "Cancelado":
            print(f"Agendamento de {self.__nome_cliente} est√° cancelado.")
        else:
            print(f"Agendamento de {self.__nome_cliente} est√° ativo para {self.data_agendamento}.")

# Subclasse com heran√ßa
class AgendamentoServicoResidencial(AgendamentoServico):
    def __init__(self, nome_cliente, tipo_servico, data_agendamento, endereco):
        super().__init__(nome_cliente, tipo_servico, data_agendamento)
        self.endereco = endereco

    def confirmar_endereco(self):
        print(f"Servi√ßo residencial ser√° realizado no endere√ßo: {self.endereco}")

# Testes com encapsulamento e l√≥gica interna
agendamento = AgendamentoServico("Ana Clara", "Limpeza", "2025-06-15")
agendamento.exibir_dados()
agendamento.verificar_agendamento()
agendamento.cancelar()
agendamento.verificar_agendamento()

# Teste dos m√©todos get e set
print("Nome atual:", agendamento.get_nome_cliente())
agendamento.set_nome_cliente("Ana Clara Silva")
print("Nome atualizado:", agendamento.get_nome_cliente())
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3

# Classe principal
class AgendamentoServico:
    def __init__(self, nome_cliente, tipo_servico, data_agendamento):
        self.__nome_cliente = nome_cliente
        self.tipo_servico = tipo_servico
        self.data_agendamento = data_agendamento
        self.status = "Agendado"

    def get_nome_cliente(self):
        return self.__nome_cliente

    def set_nome_cliente(self, novo_nome):
        if novo_nome.strip() != "":
            self.__nome_cliente = novo_nome
        else:
            print("Nome inv√°lido!")

    def exibir_dados(self):
        print(f"Cliente: {self.__nome_cliente}")
        print(f"Servi√ßo Agendado: {self.tipo_servico}")
        print(f"Data do Agendamento: {self.data_agendamento}")
        print(f"Status: {self.status}")
        print("-" * 30)

# Subclasse com polimorfismo (sobrescrevendo exibir_dados)
class AgendamentoServicoResidencial(AgendamentoServico):
    def __init__(self, nome_cliente, tipo_servico, data_agendamento, endereco):
        super().__init__(nome_cliente, tipo_servico, data_agendamento)
        self.endereco = endereco

    # Polimorfismo: sobrescrevendo o m√©todo da classe pai
    def exibir_dados(self):
        super().exibir_dados()
        print(f"Endere√ßo do Servi√ßo: {self.endereco}")
        print("-" * 30)

# ------------------------
# Integra√ß√£o com SQLite
# ------------------------

# Conectando ao banco
conexao = sqlite3.connect("agendamentos.db")
cursor = conexao.cursor()

# Criando a tabela (caso n√£o exista)
cursor.execute("""
CREATE TABLE IF NOT EXISTS agendamentos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome_cliente TEXT,
    tipo_servico TEXT,
    data_agendamento TEXT,
    endereco TEXT
)
""")
conexao.commit()

# Fun√ß√£o para inserir agendamento
def inserir_agendamento(nome, tipo, data, endereco=None):
    cursor.execute("INSERT INTO agendamentos (nome_cliente, tipo_servico, data_agendamento, endereco) VALUES (?, ?, ?, ?)", 
                   (nome, tipo, data, endereco))
    conexao.commit()
    print("Agendamento inserido com sucesso.")

# Fun√ß√£o para consultar todos os agendamentos
def consultar_agendamentos():
    cursor.execute("SELECT * FROM agendamentos")
    resultados = cursor.fetchall()
    print("=== Agendamentos no banco ===")
    for linha in resultados:
        print(linha)
    print("-" * 30)

# Fun√ß√£o para deletar agendamento por ID
def deletar_agendamento(id_agendamento):
    cursor.execute("DELETE FROM agendamentos WHERE id = ?", (id_agendamento,))
    conexao.commit()
    print(f"Agendamento com ID {id_agendamento} foi exclu√≠do.")

# -----------------------
# Testando o banco
# -----------------------
inserir_agendamento("Ana Clara", "Limpeza", "2025-06-15", "Rua das Flores, 123")
inserir_agendamento("Carlos Souza", "Manuten√ß√£o de Ar", "2025-06-10")

consultar_agendamentos()

deletar_agendamento(1)  # Excluindo o ID 1 como exemplo
consultar_agendamentos()

# Encerrando conex√£o
conexao.close()
```

## PARTE 5 ‚Äì Projeto Final e Organiza√ß√£o

**Tarefas:**
- Consolidar todas as partes do projeto em um √∫nico script organizado.
- Comentar o c√≥digo explicando cada parte.
- Testar todas as funcionalidades.

**C√≥digo:**

```python
import sqlite3

# -------------------------
# Classe base com encapsulamento
# -------------------------
class AgendamentoServico:
    def __init__(self, nome_cliente, tipo_servico, data_agendamento):
        self.__nome_cliente = nome_cliente
        self.tipo_servico = tipo_servico
        self.data_agendamento = data_agendamento
        self.status = "Agendado"

    def get_nome_cliente(self):
        return self.__nome_cliente

    def set_nome_cliente(self, novo_nome):
        if novo_nome.strip() != "":
            self.__nome_cliente = novo_nome
        else:
            print("Nome inv√°lido!")

    def remarcar(self, nova_data):
        self.data_agendamento = nova_data

    def cancelar(self):
        self.status = "Cancelado"

    def verificar_agendamento(self):
        if self.status == "Cancelado":
            print(f"Agendamento de {self.__nome_cliente} est√° cancelado.")
        else:
            print(f"Agendamento de {self.__nome_cliente} est√° ativo.")

    def exibir_dados(self):
        print(f"Cliente: {self.__nome_cliente}")
        print(f"Servi√ßo: {self.tipo_servico}")
        print(f"Data: {self.data_agendamento}")
        print(f"Status: {self.status}")
        print("-" * 30)

class AgendamentoServicoResidencial(AgendamentoServico):
    def __init__(self, nome_cliente, tipo_servico, data_agendamento, endereco):
        super().__init__(nome_cliente, tipo_servico, data_agendamento)
        self.endereco = endereco

    def exibir_dados(self):
        super().exibir_dados()
        print(f"Endere√ßo do Servi√ßo: {self.endereco}")
        print("-" * 30)

# -------------------------
# Banco de dados (SQLite)
# -------------------------
conexao = sqlite3.connect("agendamentos.db")
cursor = conexao.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS agendamentos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome_cliente TEXT,
    tipo_servico TEXT,
    data_agendamento TEXT,
    endereco TEXT
)
""")
conexao.commit()

def inserir_agendamento(nome, tipo, data, endereco=None):
    cursor.execute("INSERT INTO agendamentos (nome_cliente, tipo_servico, data_agendamento, endereco) VALUES (?, ?, ?, ?)",
                   (nome, tipo, data, endereco))
    conexao.commit()

def consultar_agendamentos():
    cursor.execute("SELECT * FROM agendamentos")
    agendamentos = cursor.fetchall()
    for ag in agendamentos:
        print(f"ID: {ag[0]} | Cliente: {ag[1]} | Servi√ßo: {ag[2]} | Data: {ag[3]} | Endere√ßo: {ag[4]}")
    return agendamentos

def deletar_agendamento(id_agendamento):
    cursor.execute("DELETE FROM agendamentos WHERE id = ?", (id_agendamento,))
    conexao.commit()

def remarcar_agendamento(id_agendamento, nova_data):
    cursor.execute("UPDATE agendamentos SET data_agendamento = ? WHERE id = ?", (nova_data, id_agendamento))
    conexao.commit()

def editar_nome_cliente(id_agendamento, novo_nome):
    cursor.execute("UPDATE agendamentos SET nome_cliente = ? WHERE id = ?", (novo_nome, id_agendamento))
    conexao.commit()

# -------------------------
# Menu Interativo
# -------------------------
def menu():
    while True:
        print("\n==== MENU DE AGENDAMENTOS ====")
        print("1. Cadastrar Agendamento")
        print("2. Listar Agendamentos")
        print("3. Editar Nome do Cliente")
        print("4. Remarcar Agendamento")
        print("5. Cancelar/Excluir Agendamento")
        print("0. Sair")
        escolha = input("Escolha uma op√ß√£o: ")

        if escolha == "1":
            nome = input("Nome do Cliente: ")
            tipo = input("Tipo de Servi√ßo: ")
            data = input("Data do Agendamento (AAAA-MM-DD): ")
            residencial = input("Servi√ßo Residencial? (s/n): ").lower()
            if residencial == "s":
                endereco = input("Endere√ßo do Servi√ßo: ")
                inserir_agendamento(nome, tipo, data, endereco)
            else:
                inserir_agendamento(nome, tipo, data)
            print("‚úÖ Agendamento inserido com sucesso.")

        elif escolha == "2":
            print("\n--- Lista de Agendamentos ---")
            consultar_agendamentos()

        elif escolha == "3":
            consultar_agendamentos()
            id_editar = input("Digite o ID do agendamento para editar: ")
            novo_nome = input("Novo nome do cliente: ")
            editar_nome_cliente(id_editar, novo_nome)
            print("‚úÖ Nome atualizado.")

        elif escolha == "4":
            consultar_agendamentos()
            id_remarcar = input("Digite o ID do agendamento para remarcar: ")
            nova_data = input("Nova data (AAAA-MM-DD): ")
            remarcar_agendamento(id_remarcar, nova_data)
            print("‚úÖ Agendamento remarcado.")

        elif escolha == "5":
            consultar_agendamentos()
            id_excluir = input("Digite o ID do agendamento para excluir: ")
            deletar_agendamento(id_excluir)
            print("‚úÖ Agendamento cancelado/exclu√≠do.")

        elif escolha == "0":
            print("Encerrando o sistema...")
            break

        else:
            print("‚ùå Op√ß√£o inv√°lida! Tente novamente.")

# -------------------------
# Execu√ß√£o do Menu
# -------------------------
menu()
conexao.close()
```
